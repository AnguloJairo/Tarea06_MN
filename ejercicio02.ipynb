```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

# Funcin original
def f(x):
    return np.arctan(x)

# Serie de Taylor centrada en x=0 hasta grado n
def taylor_arctan(x, n):
    result = 0
    for k in range(n):
        term = ((-1)**k) * (x**(2*k + 1)) / (2*k + 1)
        result += term
    return result

# Dominio de evaluacin
x_vals = np.linspace(0, 2, 400)
y_true = f(x_vals)

# --- GRFICA 1: Serie de Taylor con diferentes rdenes ---
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_true, label='arctan(x)', color='black', linewidth=2)

# Dibujamos Taylor de orden 1, 3, 5, 7
orders = [1, 3, 5, 7]
colors = ['red', 'green', 'blue', 'orange']
for n, c in zip(orders, colors):
    y_taylor = taylor_arctan(x_vals, n)
    plt.plot(x_vals, y_taylor, label=f'Taylor grado {2*n - 1}', linestyle='--', color=c)

plt.title('Serie de Taylor de arctan(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

# --- GRFICA 2: Polinomios de Lagrange con diferentes nmeros de puntos ---
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_true, label='arctan(x)', color='black', linewidth=2)

# Interpolaciones con 2, 3 y 4 puntos centrados en x=1
point_sets = [
    np.array([0.5, 1.5]),
    np.array([0.5, 1.0, 1.5]),
    np.array([0.25, 0.75, 1.25, 1.75])
]
colors = ['red', 'green', 'blue']

for pts, c in zip(point_sets, colors):
    y_pts = f(pts)
    poly = lagrange(pts, y_pts)
    y_interp = poly(x_vals)
    plt.plot(x_vals, y_interp, linestyle=':', color=c, label=f'Lagrange con {len(pts)} pts')
    plt.scatter(pts, y_pts, color=c)

plt.title('Polinomios de Lagrange para arctan(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()



```
